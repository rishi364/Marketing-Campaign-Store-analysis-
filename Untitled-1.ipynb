```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def meta_data_table(df): 
    meta_data = pd.DataFrame(columns=['Attribute', 'Description', 'Max', 'Min', 'Mean', 'Std. Deviation', 'Mode'])
    
    rows = []  
    
    for col in df.columns:
        if df[col].dtype == 'object':
            mode = df[col].mode()[0]
            rows.append({'Attribute': col, 
                         'Description': 'Nominal', 
                         'Max': None, 
                         'Min': None, 
                         'Mean': None, 
                         'Std. Deviation': None, 
                         'Mode': mode})
        else:
            rows.append({'Attribute': col, 
                         'Description': 'Numeric', 
                         'Max': df[col].max(), 
                         'Min': df[col].min(), 
                         'Mean': df[col].mean(), 
                         'Std. Deviation': df[col].std(), 
                         'Mode': df[col].mode()[0]})
    
    meta_data = pd.concat([meta_data, pd.DataFrame(rows)], ignore_index=True)
    
    return meta_data

meta_data = meta_data_table(data)
print(meta_data)

for col in data.select_dtypes(include=['float64', 'int64']).columns:
    plt.figure()
    sns.histplot(data[col], kde=True)
    plt.title(f'Histogram of {col}')
    plt.show()
```

```python
def missing_error_summary(df):
    missing_data = df.isnull().sum()
    error_data = df[df.apply(lambda x: x.astype(str).str.contains(r'[^\w\s]', regex=True).any(), axis=1)]
    
    return missing_data, error_data

missing_data, error_data = missing_error_summary(data)
print("Missing Data Summary:\n", missing_data)
print("\nError Data Sample:\n", error_data.head())
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score

data = pd.read_csv('/Users/rishisharma/Desktop/notebook-1/Marketing Campaign data.csv')

sample_data = data.sample(n=1000, random_state=1)

sample_data = sample_data.drop(columns=['CUST_ID', 'COMMENTS'])

sample_data = pd.get_dummies(sample_data, drop_first=True)

X = sample_data.drop(columns=['AFFINITY_CARD'])
y = sample_data['AFFINITY_CARD']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

model = LogisticRegression(max_iter=1000)
model.fit(X_train_scaled, y_train)

y_pred = model.predict(X_test_scaled)

print("Accuracy Score:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
def preprocess_input(input_data, label_encoders, target_encoder):
    processed_data = {}
    for column, value in input_data.items():
        if column in label_encoders:
            le = label_encoders[column]
            if value not in le.classes_:
                print(f"Value '{value}' not found in encoder classes for column '{column}'.")
                return None
            processed_data[column] = le.transform([value])[0]
        else:
            processed_data[column] = value
    return pd.DataFrame([processed_data])

def predict(input_data):
    X_input = preprocess_input(input_data, label_encoders, target_encoder)
    if X_input is None:
        return "Error in input data"
    prediction = model.predict(X_input)
    return target_encoder.inverse_transform(prediction)[0]
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import joblib

data_cleaned = pd.read_csv('Marketing Campaign data.csv')

label_encoders = {}
for column in data_cleaned.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    data_cleaned[column] = le.fit_transform(data_cleaned[column].astype(str))
    label_encoders[column] = le

X = data_cleaned.drop(columns=['AFFINITY_CARD'])
y = data_cleaned['AFFINITY_CARD']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)

joblib.dump(dt_model, 'decision_tree_model.pkl')
joblib.dump(label_encoders, 'label_encoders.pkl')

from sklearn.metrics import accuracy_score
y_pred = dt_model.predict(X_test)
print(f'Decision Tree Accuracy: {accuracy_score(y_test, y_pred)}')
```
